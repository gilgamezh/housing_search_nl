```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_json("../scrape_funda/scraped_amsterdam.json")
```

```python

df["construction_year"] = df['construction_year'].str.extract(pat='([0-9]{4})').astype(float)
df['price_per_living_sqm'] = df['price']/df['sqm']
```
